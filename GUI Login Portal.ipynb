```python
from tkinter import *
from PIL import ImageTk,Image
from tkinter import messagebox


def handle_login():
    email=email_input.get()
    password=password_input.get()
    if email=='dheeraj.tanwar@gmail.com' and password=="1234":
        messagebox.showinfo("Login Portal","Login Successful")
    else:
        messagebox.showerror("Login Portal","Login Error")


root=Tk()

root.title("Dheeraj Portal")
root.minsize(350,500)
root.maxsize(400,500)

root.configure(background='black')
root.iconbitmap('photo.ico')


img=Image.open('photo1.png')
resize_img=img.resize((120,120))
img=ImageTk.PhotoImage(resize_img)


img_label=Label(root,image=img)
img_label.pack(pady=(10,10))


text_label=Label(root,text="Welcome Dheeraj",bg='black',fg='white')
text_label.pack()
text_label.configure(font=("Algerian",20))


email_label=Label(root,text="Email Login",bg='black',fg='white')
email_label.pack(pady=(20,1))
email_label.configure(font=("Arial Narrow",14))

email_input=Entry(root,bg='red',width=50)
email_input.pack(ipady=2)


password_label=Label(root,text="Password Here",bg='black',fg='white')
password_label.pack(pady=(15,1))
password_label.configure(font=("Arial Narrow",14))

password_input=Entry(root,bg='red',width=50)
password_input.pack(ipady=2)


login_bttn=Button(root,text='Login Here',bg='white',fg='black',width=18,height=1,command=handle_login)
login_bttn.pack(pady=(30,10))
login_bttn.config(font=("Arial Rounded MT Bold",14))



mainloop()
```
